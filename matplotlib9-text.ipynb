```python
## 10. 그래프에 text 추가
'''

    plt.text(x,y, mesg ) 

'''
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
# ax = plt.axes()
x = np.arange(3)  # [0, 1, 2]
values = [100, 400, 900]

plt.bar(x, values)
plt.axhline(200, color='r', linestyle="--")
plt.xticks(x, labels=['2017', '2018', '2019'])

plt.text(0, 800, 'hello', fontsize = 'xx-large', fontstyle='italic')
# text 지정
for idx, v in enumerate(values):
    # ax.text(x[idx], v, v)  # ax.text(x,y, mesg )
    plt.text(x[idx], v, v )  # plt.text(x,y, mesg )
plt.show()
```
