```python
# 2. 다중 그래프

# 가) fig.add_subplot(행,열, n번째열)
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)  # 1행 2열의 1번째열
plt.plot([1,2,3,4])  # ax1.plot([1,2,3,4]) 대신에 plt 사용
plt.title("sub1")

ax2= fig.add_subplot(1,2,2)   # 1행 2열의 2번째열
plt.plot([-1,-2,-3,-4])
plt.title("sub2")

fig.suptitle("Main")
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)  # 1행 1열의 1번째열
plt.plot([1,2,3,4])  # ax1.plot([1,2,3,4]) 대신에 plt 사용
plt.title("sub1")

ax2= fig.add_subplot(2,2,2)   # 1행 1열의 2번째열
plt.plot([-1,-2,-3,-4])
plt.title("sub2")

ax3 = fig.add_subplot(2,2,3)  # 2행 1열의 1번째열
plt.plot([1,2,3,4])  # ax1.plot([1,2,3,4]) 대신에 plt 사용
plt.title("sub3")

ax4= fig.add_subplot(2,2,4)   # 2행 2열의 2번째열
plt.plot([-1,-2,-3,-4])
plt.title("sub4")

fig.suptitle("Main")
plt.show()
```

```python
# 나) plt.subplots(행,열)
import matplotlib.pyplot as plt
fig, axs = plt.subplots(nrows=1, ncols=2) # 1행 2열
print(axs[0])
axs[0].plot([1,2,3])
axs[1].plot([-1,-2,-3])
plt.show()
```
