```python
# 9. box

## 가) 기본
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(1234)
# 박스plot는 홀수값 지정
x = np.random.random(9)
print(x)
'''
[0.9581393536837052, 0.8018721775350193, 
0.7853585837137692,  # ( 3사분위 )
0.7799758081188035, 
0.6221087710398319, # 중앙값 ( 2사분위 ) 
0.4377277390071145, 
0.2764642551430967,  # ( 1사분위 ) 
0.2725926052826416, 0.1915194503788923]
'''
print(sorted(x, reverse=True))
plt.boxplot(x,  labels=['xx_value'])
plt.show()
```

```python
print("1사분위:", np.quantile(x, 0.25)) # 0.2764642551430967
print("2사분위:", np.quantile(x, 0.5))  # 0.6221087710398319
print("3사분위:", np.quantile(x, 0.75)) # 0.7853585837137692
print("IQR: ", np.quantile(x, 0.75) - np.quantile(x, 0.25))
```

박스 플롯 너머 있는 선: 최대 = IQR(4분위 범위) * 1.5 + Q3

박스 플롯 윗 선: Q3(3사분위) = 75%
박스 플롯 중간에 빨간줄: Q2(2사분위) = 중앙값 50%
박스 플롯 아래 선: Q1(1사분위) = 25%

박스 플롯너머 아래 있는 선: 최소 = IQR(4분위 범위) * 1.5

```python
plt.boxplot([x,x,x],  labels=['xx1_value', 'xx2_value', 'xx3_value'])  # 여러 데이터 비교 가능
plt.show()
```
